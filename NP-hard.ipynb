# Informe Problemas NP-Hard

## ¿ Qué quiere decir que un problema **NP-Hard**?

Un problema **NP-hard** corresponde a un tipo de problema **NP** (los mas dificiles en la practica ), un problema **NP** es aquel que no puede ser resuelto en un tiempo razonable , esto asumiendo la conjetura de que $P\neq NP$.

Aun asi estos problemas pueden ser reducidos a otros problemas del tipo **NP**, a pesar de no poder encontrar una solucion en un tiempo polinomial, la verificacion de una posible solucion a estos **NP**, si puede serlo.

---

## ¿ Qué quiere decir que $P\neq NP$ ?

Primero definiremos que significa que un problema sea P o NP:


los problemas **P**:
*   Un problema entra en la categoria de P , cuando este puede ser resuelto con un algoritmo en tiempo polinomial en funcion de la entrada.

En cambio los problemas **NP**
*   Como anteriormente mencione , una posible solucion al problema obtenida de alguna forma, es posible realizar una verificacion de esta en tiempo polinomial. 



Esta es una conjetura dentro de la complejidad computacional, se cree que $P\neq NP$ es cierta, ya que si fuera del caso contrario todo problema computacional podria ser resuelto en un tiempo **polinomial**.



![image](https://imgur.com/N9LqMhW.png)

Los problemas P es una subcase de NP, más no todo problema del tipo NP será necesariamente posible de resolver en tiempo polinomial.

---

## ¿ Cómo se prueba que un problema es NP-Hard usando reducciones?

Para demostrar que un problema **B** es **intratable** o **NP-Hard** debemos aplicar ***Reducción***:



1. Esto quiere decir que seleccionamos un problema **A** que sabemos que es **NP-Hard**.
2. Si el problema **A** es **reducible** a **B**, entonces el problema **B**  es **NP-Hard** también.



Cualquier problema $A$ se reduce a otro problema $B$ si un algoritmo que resuelve a $B$ puede ser transformado en otro que resuelve a $A$. Esta transformación se refiere a que $A$ puede ser resuelto usando una cantidad polinomial de llamadas a una subrutina que resuelve al problema $B$, además del trabajo adicional.


![image](https://i.imgur.com/fQcjKgp.png)



Además, si un problema $A$ se puede reducir a un problema $B$ y $B$ es resuelto en un tiempo polinomial, entonces $A$ también. 



![image](https://i.imgur.com/ISSYWYD.png)


Esta observación se puede utilizar para comprobar que un problema es del tipo **NP-Hard**

# A) **Clique máximo**

#### Descripción del problema: 

---



>**Entrada** : Dado un grafo no dirigido $G = (V,E)$ con $V$ vértices y $E$ arcos
>
>**Salida** : Un *clique* de $G$ con la mayor cantidad de elementos posibles

Se define *clique* como un subconjunto de vértices de un grafo sin dirección de forma tal que cada par de vértices distintos en el *clique* son adjacentes.

Este problema consiste en encontrar el conjunto de vértices que conforman el clique de mayor tamaño posible dentro del grafo $G$

</center>
Para nuestro problema $A$ utilizaremos nuevamente el problema del `Conjunto independiente' en su versión de desición

### **Prueba de NP-Hard**

#### **Teorema 2 (NP-Hardness de Clique**)
El problema *Clique* es *NP-Hard*



Para probar que este problema pertenece al conjunto **NP-Hard** es necesario escoger algún otro problema $A$ y demostrar que $A$ se reduce a nuestro problema objetivo, al cual denominaremos $B$

Siendo $A$ **NP-Hard** y asumiendo que la conjetura $P \neq NP$ es cierta, $B$ también debe ser **NP-Hard** puesto a que cualquier otro caso se opondría a la conjetura.

Para nuestro problema $A$ utilizaremos nuevamente el problema del ***Conjunto independiente máximo*** 

---
#### **1) Selección del problema $A$**

*Descripción del problema : Conjunto Independiente Máximo*

>***Entrada***: Grafo no dirigido $G=(V,E)$ con $V$ vértices y $E$ arcos
>
>***Salida***: Conjunto independiente con la cantidad máxima de nodos




![image](https://imgur.com/wQdrBqn.png)


En la imagen superior se encuentran algunos ejemplos de conjuntos independientes dados sus grafos correspondientes. Se nota que para que un conjunto sea independiente, los nodos que lo conforman no deben ser adyacentes entre si



Ya que queremos demostrar que nuestro problema $B$ *(Clique máximo)* es del tipo **NP-Hard**, debemos reducir efectivamente el problema $A$ a $B$

---

#### **2)Reducción**

Igual que en el caso anterior, lo primero que hay que hacer es demostrar que el problema del conjunto independiente se puede reducir a *clique*.

Como es habitual, el algoritmo que aprovecha esta reducción sigue el siguiente esquema : 

![image](https://i.imgur.com/L6yMMud.png)

#### **Lema 1**

El problema del Conjunto independiente máximo se reduce a Clique máximo

#### **Prueba**

Se puede observar que para que un conjunto sea independiente, cada vértice dentro de este no debe tener ningún arco hacia ningún otro.

Por otro lado, todos los nodos en un *clique* están conectados entre si.

Es fácil observar como estas dos estructuras se encuentran conectadas. Un conjunto independiente es un *clique* en el complemento del grafo de entrada $G$ y viceversa. Esto se debe a que si un conjunto de nodos se encuentran todos conectados entre sí, su complemento tendrá $0$ arcos.

La siguiente imagen representa esta asociación a través de un ejemplo :

<img src='https://i.imgur.com/BxQveXo.png'>

Se nota que el conjunto independiente de tamaño $3$ en el grafo complementario $G'$ corresponde a un *clique* para el grafo original $G$.

Con tal de encontrar el conjunto independiente máximo, solo basta con encontrar el clique máximo de un grafo $G$ y retornar su complemento.

Por lo tanto, el problema del *conjunto independiente máximo* puede ser reducido al *clique máximo* de un grafo $G$

# B) The salesman problem (TSP)

El problema consiste en encontrar la ruta más corta posible que visite cada ciudad exactamente una vez y vuelva al punto de partida.

![image](https://imgur.com/fdKirIX.png)



>**Entrada** : Un grafo completo no dirigido G=(V,E) y un valor Ce para cada arco e ∈ E .
>
>**Salida** :Un tour T ⊆ E de G con el coste mínimo, es decir, que minimice  ∑Ce   (sumatoria de los arcos).

---
### **Prueba de NP-Hard**

#### **Teorema 2 (NP-Hard de TSP**)
El problema *TSP* es *NP-Hard*



Para probar que este problema pertenece al conjunto **NP-Hard** es necesario escoger algún otro problema $A$ y demostrar que $A$ se reduce a nuestro problema objetivo, al cual denominaremos $B$

Siendo $A$ **NP-Hard** y asumiendo que la conjetura $P \neq NP$ es cierta, $B$ también debe ser **NP-Hard** puesto a que cualquier otro caso se opondría a la conjetura.

Para nuestro problema $A$ utilizaremos nuevamente el problema del ***Ciclo hamiltoniano***.

---
#### **1) Selección del problema $A$**

Descripcion problema $A$ ***Ciclo hamiltoniano***.

>**Entrada**: Un grafo dirigido $G=(V,E)$ con un nodo inicial $s \in V$ y un nodo final $t \in V$.
>
>**Salida**: “SI”, si $G$ contiene un camino que parte en $s$ y termina en $t$, pasando por todos los nodos en $V$. “NO”, si no existe el camino.



Como ejemplo, en la siguiente imagen, podemos ver que en el grafo de la izquierda **SI existe un ciclo Hamiltoniano** (ruta de las flechas punteadas). Sin embargo, en el grafo de la izquierda, la respuesta es NO.

![image](https://imgur.com/fUY15OA.png)

---

#### **2)Reduccion**

*Reducción a ciclo hamiltoniano:*

Para realizar la reduccion tomaremos un grafo y eliminaremos todas las aristas. luego tomaremos un vértice arbitrario $v$ y volveremos a añadir aristas dirigidas a partir de ese vértice si el destino de la arista añadida no está más lejos de $v$ que $k$ la constante de longitud del problema de decisión tsp.

Si el grafo construido tiene un ciclo hamiltoniano entonces también es un recorrido de longitud < k en el grafo original, por lo tanto una solución al TSP.

A la inversa, si no hay ningún ciclo hamiltoniano en el grafo: como se han construido todos los recorridos de longitud < k que empiezan en el vértice v y como una solución al TSP es un recorrido de longitud < k que empieza en v, y un ciclo hamiltoniano, no hay solución al TSP en el grafo original.
